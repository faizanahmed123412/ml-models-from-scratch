```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1a: Generate Synthetic Data ---

# Set a seed for reproducibility
np.random.seed(0)

# Generate data for 50 students who failed (class 0)
# They studied between 0 and 4.5 hours (on average)
X_failed = np.random.normal(2.5, 1, 50)
y_failed = np.zeros(50, dtype=int) # Label them as 0
print(X_failed)

# Generate data for 50 students who passed (class 1)
# They studied between 3.5 and 8 hours (on average)
X_passed = np.random.normal(6, 1, 50)
y_passed = np.ones(50, dtype=int) # Label them as 1
print(X_passed)

# Combine the data into single variables
# We use .reshape(-1, 1) to make X a column vector
X = np.concatenate((X_failed, X_passed)).reshape(-1, 1)
y = np.concatenate((y_failed, y_passed)).reshape(-1, 1)
print(X)
print(y)

# --- 1b: Visualize the Data ---
plt.figure(figsize=(10, 6))

# Plot the 'failed' students in red
plt.scatter(X[y.flatten() == 0], y[y.flatten() == 0], c='red', marker='x', label='Failed (0)')

# Plot the 'passed' students in blue
plt.scatter(X[y.flatten() == 1], y[y.flatten() == 1], c='blue', marker='o', label='Passed (1)')

plt.title('Exam Pass/Fail vs. Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Outcome (0 or 1)')
plt.yticks([0, 1]) # Make sure the y-axis only shows 0 and 1
plt.legend()
plt.grid(True)
plt.show()
```
